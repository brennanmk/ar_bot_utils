```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("~/ray_results/PPO_2024-03-05_13-20-52/PPO_ARBotTabletopGym_1acdb_00000_0_2024-03-05_13-20-54/progress.csv")
df2 = pd.read_csv("~/ray_results/PPO_2024-03-05_13-30-07/PPO_ARBotTabletopGym_643a4_00000_0_2024-03-05_13-30-07/progress.csv")
df3 = pd.read_csv("~/ray_results/PPO_2024-03-05_13-39-30/PPO_ARBotTabletopGym_b3ed1_00000_0_2024-03-05_13-39-30/progress.csv")
```

```python
frames = np.array([df1["episode_reward_mean"], df2["episode_reward_mean"], df3["episode_reward_mean"]])

plt.errorbar(x=np.arange(frames.shape[1]),y=frames.mean(0),yerr=frames.std(0),ecolor='lightblue')

plt.legend()
plt.ylabel("Average Episode Reward")
plt.xlabel("Training Iteration")
plt.title("Time To Train ARBot with No Obstacles")

plt.show()
```
